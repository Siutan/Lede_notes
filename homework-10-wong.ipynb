```python
url = "https://api.forecast.io/forecast/0d252b648d0afdebfc11fae57d900f9b/22.2783,114.1747"
```

```python
response = urllib2.urlopen(url)
```

```python
contents = response.read()
contents
```

```python
import json
```

```python
data = json.loads(contents)
data
```

```python
data.keys()
```

```python
# What is the current wind speed?
data['currently']['windSpeed']
```

```python
tomorrow_forecast = data['daily']['data'][1]
tomorrow_forecast
```

```python
# How much of the moon will be visible tomorrow?
tomorrow_forecast['moonPhase']
```

```python
# What will be the swing between the high and low temperature?
float(tomorrow_forecast['temperatureMax']) - float(tomorrow_forecast['temperatureMin'])
```

```python
# Convert the UNIX timestamp 1435601280 into a datetime using Python (Google is your buddy)
```

```python
import datetime
```

```python
#http://www.avilpage.com/2014/11/python-unix-timestamp-utc-and-their.html
converted = datetime.datetime.utcfromtimestamp(1435601280)
```

```python
converted.strftime("%Y %B %-d %H:%M")
```
